## Create dimension tables

```python
dimensions_joined = metadata.merge(dimensions,how="right", on=["table_schema","table_name","column_name"])
dim_names = list(set(dimensions["dim_name"]))
```

```python
for name in dim_names:
    sql = ["CREATE TABLE {}(\n skey serial primary key,\n".format(name)]
    
    for row in dimensions_joined[dimensions_joined["dim_name"] == name].to_numpy():
        col_name = row[3]
        is_nullable = "" if row[6]=="YES" else " NOT NULL"
        data_type = row[27] # udt_name column in the metadata
        character_maximum_length = row[8]
        
        if (data_type in ["char", "varchar"]):
            sql_row = "{} {}({}){},\n".format(col_name, data_type, int(character_maximum_length), is_nullable)
        else:
            sql_row = "{} {}{},\n".format(col_name, data_type, is_nullable)
        
        sql.append(sql_row)
    
    sql = "".join(sql)[:-2]+"\n)\n"
    print(sql)
        
        
        
```
